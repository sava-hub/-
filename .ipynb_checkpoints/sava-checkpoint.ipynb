```python
import numpy as np
import scipy as sp
import scipy.stats
import matplotlib.pyplot as plt
rv = sp.stats.norm(loc=0, scale=1)
```

```python
#matplotlib에 남아있는 font 캐시 삭제 
plt.rcParams["font.family"] = '나눔고딕'
plt.rcParams['axes.unicode_minus'] = False
print (plt.rcParams['font.size'] ) 
print (plt.rcParams['font.family'] )
```

```python
xx = np.linspace(-8, 8, 100)
cdf = rv.cdf(xx)
plt.plot(xx, cdf)
plt.title("Density Function")
plt.xlabel("$x$")
plt.ylabel("$F(x)$")
plt.show()
```

```python
print(rv.ppf(373/380)) #7등 분위수값
print(rv.ppf(20/380))
print(rv.ppf(353/360))
print(rv.ppf(120/360)) 
```

```python
# 평균 +  분산*2.0875094772894123= 1446.7
# 평균 +  분산*-1.6198562586382699= 450
A = np.array([[1,2.0875094772894123],[1,-1.6198562586382699]])
B = np.array([1446.7, 450])
C = np.linalg.solve(A, B)
print(C)
```

```python
#낙타분포
# 평균 +  분산*2.0653606500222765= 1446.7
# 평균 +  분산*-0.43072729929545756
A = np.array([[1,2.0653606500222765],[1,-0.43072729929545756]])
B = np.array([1446.7, 811.1])
C2 = np.linalg.solve(A, B)
print(C2)
```

```python
rv2 = sp.stats.norm(loc=C[0], scale=C[1])
xx = np.linspace(0, 1500, 15000)
pdf = rv2.pdf(xx)
plt.plot(xx, pdf)
plt.title("Density Function")
plt.xlabel("$x$")
plt.ylabel("$F(x)$")
plt.show()
```

```python
rv3 = sp.stats.norm(loc=C[0], scale=C[1])
xx = np.linspace(0, 1500, 15000)
cdf = rv3.cdf(xx)
plt.plot(xx, cdf)
plt.title("Density Function")
plt.xlabel("$x$")
plt.ylabel("$F(x)$")
plt.show()
```

```python
#과목 평균 등급컷
grade = np.array([0.04, 0.11, 0.23, 0.40, 0.60, 0.77, 0.89, 0.96])
grade = np.array([0.96, 0.89, 0.77, 0.60, 0.40, 0.23, 0.11, 0.04])

for i in grade:
    print(np.round(rv2.ppf(i)/15, 2))
```

```python
#낙타분포 등급컷
rv4 = sp.stats.norm(loc=C2[0], scale=C2[1])
grade = np.array([0.04, 0.11, 0.23, 0.40, 0.60, 0.77, 0.89, 0.96])
grade = np.array([0.96, 0.89, 0.77, 0.60, 0.40, 0.23, 0.11, 0.04])

for i in grade:
    print(np.round(rv4.ppf(i)/15, 2))
```
